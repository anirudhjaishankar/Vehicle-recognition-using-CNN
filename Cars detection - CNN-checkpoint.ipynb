```python
#Importing and initialisations
import pandas as pd
import numpy as np
import cv2  
target_nvehi = np.zeros(1000)
target_vehi = np.ones(1000)
```

```python
import matplotlib.pyplot as plt
im = cv2.imread('Cars dataset/images/non-vehicles/Far/image0000.png')
imv = image = cv2.imread('Cars dataset/images/vehicles/Far/image0000.png')
im = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
hist = cv2.calcHist([im],[0],None,[256],[0,256])
histv = cv2.calcHist([imv],[0],None,[256],[0,256])
plt.plot(hist,color='blue')
plt.plot(histv,color='red')
plt.show()
```

```python
#Pre processing
# for non vehicle images
predictors_nvehi = [] 
for i in range(0,1000):
    a = "{0:04d}".format(i)
    image = cv2.imread('Cars dataset/images/non-vehicles/Far/image'+a+'.png')
        #Finding HOG
#     hog = cv2.HOGDescriptor()
#     h = hog.compute(gray_img)

    #Converting to grayscale
    gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    gray_image = cv2.GaussianBlur(gray_image,(3,3),4)
    gray_image = [(abs(i-np.mean(gray_image)))/255 for i in gray_image]
    #Creating the predictor array
#     gray_image = gray_image.flatten()
    predictors_nvehi.append(gray_image)    
predictors_nvehi = np.array(predictors_nvehi)    
print(predictors_nvehi.shape)
print(predictors_nvehi[:3,:])
```

```python
#Pre processing
#for vehicle images
predictors_vehi = []
for i in range(0,1000):
    a = "{0:04d}".format(i)
    image = cv2.imread('Cars dataset/images/vehicles/Far/image'+a+'.png')
        #Finding HOG
#     hog = cv2.HOGDescriptor()
#     h = hog.compute(gray_img)

    #Converting to grayscale
    gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    gray_image = cv2.GaussianBlur(gray_image,(3,3),4)
    gray_image = [(abs(i-np.mean(gray_image)))/255 for i in gray_image]
    #Creating the predictor array
#     gray_image = gray_image.flatten()
    predictors_vehi.append(gray_image)    
predictors_vehi = np.array(predictors_vehi)
print(predictors_vehi.shape)
print(predictors_vehi[:3,:])
```

```python
from keras.utils import to_categorical
targets = np.append(target_nvehi,target_vehi,axis=0)
predictors = np.append(predictors_nvehi,predictors_vehi,axis=0)
# total = np.column_stack((predictors,targets))
# np.random.shuffle(total)
# predictors = total[:,:-1]
# target = total[:,-1]
print(predictors[-1,:])
target = to_categorical(targets)
predictors = predictors.reshape(-1,64,64,1)
print(predictors.shape,target.shape)
print(target[-1,])
```

```python
import keras
from keras.layers import Conv2D,Flatten,Dense,MaxPool2D
from keras.models import Sequential
from keras.callbacks import EarlyStopping,ModelCheckpoint

model = Sequential()

model.add(Conv2D(32,kernel_size=3,activation='relu',input_shape=(64,64,1),padding='same'))
model.add(MaxPool2D(2))
model.add(Conv2D(20,kernel_size=3,activation='relu'))
model.add(MaxPool2D(3))
model.add(Conv2D(10,kernel_size=3,activation='relu'))


model.add(Flatten())
model.add(Dense(2,activation='softmax'))

model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
train = model.fit(predictors,target,validation_split=0.3,epochs=20,callbacks=[EarlyStopping(patience=3),ModelCheckpoint('cars_weights.hdf5',monitor='val_loss',save_best_only=True)])

model.save("cars_model.h5")
```

```python
import matplotlib.pyplot as plt
history = train.history
print(history.keys())
plt.plot(history['loss'])
plt.plot(history['val_loss'])
plt.show()
```
